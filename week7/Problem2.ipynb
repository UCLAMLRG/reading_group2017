A convicted criminal who reoffends after release is known as a recidivist. The table below lists a dataset that describes prisoners released on parole, and whether they reoffended within two years of release.

This dataset lists six instances where prisoners were granted parole. Each of these instances are described in terms of three binary descriptive features (GOOD BEHAVIOR, AGE < 30, DRUG DEPENDENT) and a binary target feature, RECIDIVIST. The GOOD BEHAVIOR feature has a value of true if the prisoner had not committed any infringements during incarceration, the AGE â€  30 has a value of true if the prisoner was under 30 years of age when granted parole, and the DRUG DEPENDENT feature is true if the prisoner had a drug addiction at the time of parole. The target feature, RECIDIVIST, has a true value if the prisoner was arrested within two years of being released; otherwise it has a value of false.

<b>a. Using this dataset, construct the decision tree that would be generated by the ID3 algorithm, using entropy-based information gain. </b>


```python
import numpy as np
import math

goodBahavior=[False,False,False,True,True,True]
age30=[True,False,True,False,False,False]
drugDependent=[False,False,False,False,True,False]
recidivist=[True,False,True,False,True,False]
```

First, we will begin by calculating the total entropy of the recidivist. Notice that there are 3 True and 3 False, thus we can calculated the value as follows.

```python
entropyTotal=(3/6*math.log2(3/6)*2)*-1
```

In order to use the ID3 code we must calculate the Information gain of each of the features. 

```python
partitionGoodTrue=(2/3*math.log2(2/3)+1/3*math.log2(1/3))*-1
partitionGoodFalse=(2/3*math.log2(2/3)+1/3*math.log2(1/3))*-1
remGood=3/6*partitionGoodTrue+3/6*partitionGoodFalse
IGGood=entropyTotal-remGood

partitionAgeTrue=(2/2*math.log2(2/2))*-1
partitionAgeFalse=(3/4*math.log2(3/4)+1/4*math.log2(1/4))*-1
remAge=2/6*partitionAgeTrue+4/6*partitionAgeFalse
IGAge=entropyTotal-remAge

partitionDrugTrue=(1/1*math.log2(1/1))*-1
partitionDrugFalse=(2/5*math.log2(2/5)+3/5*math.log2(3/5))*-1
remDrug=1/6*partitionDrugTrue+5/6*partitionDrugFalse
IGDrug=entropyTotal-remDrug

print("IG Age < 30 = "+str(IGAge)+"\n"+"IG Good Bevahior = "+str(IGGood)+"\n"+"IG Drug Dependency = "+str(IGDrug)+"\n")
```

We can see that the information gain for the age < 30 feature has the highest value. This will now be the root node and we shall split the data into true and false catolgories from this node. 

Let's now look at the age < 30 true data.


```python
entropyTotal=(2/2*math.log2(2/2)*1)*-1

partitionGoodTrue=0
partitionGoodFalse=(2/2*math.log2(2/2))*-1
remGood=0*partitionGoodTrue+2/2*partitionGoodFalse
IGGood=entropyTotal-remGood

partitionDrugTrue=0
partitionDrugFalse=(2/2*math.log2(2/2))*-1
remDrug=0*partitionDrugTrue+2/2*partitionDrugFalse
IGDrug=entropyTotal-remDrug
print("IG Good Bevahior = "+str(IGGood)+"\n"+"IG Drug Dependency = "+str(IGDrug)+"\n")
```

Because all outcomes from this data provide a true recidivist, we cannot break this branch any further. This now becomes a leaf node (terminal node) for the desicion tree. All subjects age < 30 will automatically be catagolorized as true recidivist. All other features will be ignored in this branch. 

Let's now look at the age < 30 false branch.

```python
entropyTotal=(3/4*math.log2(3/4)+1/4*math.log2(1/4))*-1

partitionGoodTrue=(2/3*math.log2(2/3)+1/3*math.log2(1/3))*-1
partitionGoodFalse=(1/1*math.log2(1/1))*-1
remGood=3/4*partitionGoodTrue+1/4*partitionGoodFalse
IGGood=entropyTotal-remGood

partitionDrugTrue=1/1*math.log2(1/1)*-1
partitionDrugFalse=(3/3*math.log2(3/3))*-1
remDrug=1/4*partitionDrugTrue+3/4*partitionDrugFalse
IGDrug=entropyTotal-remDrug

print("IG Good Bevahior = "+str(IGGood)+"\n"+"IG Drug Dependency = "+str(IGDrug)+"\n")
```

We can see that the information gain for drug denpendency is the highest, thus it will provide the next branch on our desicion tree. Since only one feature is left, we can assume good behavior will be our leaf node for this branch.

Thus, our tree is now complete.
